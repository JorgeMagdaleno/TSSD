```python
filename = "customers-100.csv"

fields = []
rows = []

with open(filename, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    fields = next(csvreader)
    for row in csvreader:
        rows.append(row)
    print("Rows totales = %d" % (csvreader.line_num))

print("Los encabezados son:" + ", ".join(field for field in fields))
print('Las primeras lineas son:')

for row in rows[:5]:
    for col in row:
        print("%10s" % col, end= " "),
    print("\n")
```

```python
def has_header(filename):
    with open(filename, 'r', newline='',encoding='utf-8') as file:
        content = file.read(1024)
        sniffer = csv.Sniffer()
        return sniffer.has_header(content)

if has_header(filename):
    print("Tiene encabezados")
else:
    print("No tiene encabezados")
```

```python
def explorar_y_escribir_json(input_filename, output_filename):
    # Leer json
    with open(input_filename, 'r', encoding='utf-8') as file:
        data = json.load(file)

    # Explorar y mostrar el contenido
    print(f"Explorando archivo JSON: {input_filename}")
    explorar_json(data)

    # Escribir el resultado en un nuevo archivo
    with open(output_filename, 'w', encoding='utf-8') as file:
        file.write(json.dumps(data))
    
    print(f"Contenido explorado guardado en: {output_filename}")

def explorar_json(data, nivel=0):
    indent = '  ' * nivel 
    # Si es diccionario
    if isinstance(data, dict):
        print(f"{indent}Objeto:")
        for key, value in data.items():
            print(f"{indent}  Clave: {key}")
            explorar_json(value, nivel + 1)
    #Si es lista
    elif isinstance(data, list):
        print(f"{indent}Lista con {len(data)} elementos:")
        for index, item in enumerate(data):
            print(f"{indent}  Elemento {index}:")
            explorar_json(item, nivel + 1)
    #Para el resto de datos
    else:
        print(f"{indent}Valor: {data}")

explorar_y_escribir_json('Example_2.json', 'Example_2_output.json')
explorar_y_escribir_json('Example_4.json', 'Example_4_output.json')
explorar_y_escribir_json('Example_5.json', 'Example_5_output.json')
```
